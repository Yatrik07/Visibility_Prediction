```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

```python
df=  pd.read_csv("SData.csv")
```

```python
df
```

```python
df.drop("VISIBILITY",axis=1)
```

```python
model = RandomForestRegressor(n_estimators= 500, n_jobs= -1 , verbose=3)
```

```python
model.fit(df.drop("VISIBILITY",axis=1) , df['VISIBILITY'])
```

```python
import numpy as np
```

```python
pred = model.predict(df.drop("VISIBILITY",axis=1))
```

```python
pred = model.score(df.drop("VISIBILITY",axis=1) , df['VISIBILITY'] )
```

```python
pred
```

```python
import joblib
joblib.dump(model , "FINAL_RANDOMFOREST.pickle")
```
